```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import math
```

```python
x = [2, 4]
print(x)

# passo 1: quadrado
quadrado = [x[0]**2, x[1]**2]
print(quadrado)

# passo 2: média
media = sum(quadrado) / len(quadrado)
print(media)

# passo 3: raiz
import math
rms = math.sqrt(media)
print(rms)

# passo 4: normalizar
norm = [x[0]/rms, x[1]/rms]
print(norm)

# passo 5: peso treinável
peso = [1, 1]
print(peso)

# multiplicação elemento a elemento
saida = [norm[i] * peso[i] for i in range(len(norm))]

print(saida)
```

```python
import numpy as np

x = np.array([2, 4])

rms = np.sqrt(np.mean(x**2))
norm = x / rms

peso = np.array([1, 1])
saida = norm * peso

print(saida)
```

```python
import torch

# vetor
x = torch.tensor([2.0, 4.0])

# passo 1: RMS
rms = torch.sqrt(torch.mean(x**2))

# passo 2: normalizar
norm = x / rms

# passo 3: peso (parâmetro treinável em redes reais)
peso = torch.tensor([1.0, 1.0])

# passo 4: saída final
saida = norm * peso

print(saida)
```
